```python
# Use jellyfish to do K-mer analysis

# We now need to perform K-mer analysis with jellyfish in the server environment, so we have to confirm whether jellyfish is installed o your server.

findtool jellyfish

# If jellyfish is installed on your server, the following content will be displayed.
▶ 软件: jellyfish    环境: assembly
   路径: /opt/envs/assembly/bin/jellyfish
   版本: jellyfish 2.2.10
   激活:
     source /opt/miniforge3/etc/profile.d/conda.sh
     conda activate "/opt/envs/assembly"
   使用:
     jellyfish  [你的参数...]

▶ 软件: jellyfish    环境: transcriptome
   路径: /opt/envs/transcriptome/bin/jellyfish
   版本: jellyfish 2.2.10
   激活:
     source /opt/miniforge3/etc/profile.d/conda.sh
     conda activate "/opt/envs/transcriptome"
   使用:
     jellyfish  [你的参数...]

# We can see there are two versions on my server, and we can choose either one to activate the conda environment (I take assembly as an example)

source /opt/miniforge3/etc/profile.d/conda.sh
conda activate "/opt/envs/assembly"
which jellyfish
# The which command will display the path of jellyfish.
jellyfish --version

# The following content will be displayed
jellyfish 2.2.10

 # If not, we need to install it. We choose to install jellyfish locally from GitHub

 wget https://github.com/gmarcais/Jellyfish/releases/download/v.2.3.0/jellyfish-2.3.0

 # Unzip

 tar -xbz jellyfish-2.3.0.tar.gz

 # Enter the directory

 cd jellyfish-2.3.0/

 # Configure environment variables. Compile and install
 
 ./configure --prefix=$HOME
 make -j 4
 make install
 jellyfish --version


# After configuration, we can proceed with K-mer analysis.
# Count is the command used by jellyfish to implement the K-mer function.

jellyfish count -m 21 -s 20G -t 16 -C \
-o my_kmer21.jf \
<(zcat ~/survey/00_Clean/shi_N02_hifi_ye.clean.R1.fq.gz) \
<(zcat ~/survey/00_Clean/shi_N02_hifi_ye.clean.R2.fq.gz)
# Set the desired parameters in the count command.
# Here we take the whole genime of Mallotus repandus (2N) as an example and we set the K-value to 21.
# The name of my output file: my_kmer21.jf

# View partial statistical information
jellyfish stats my_kmer21.jf

# An example of reasonable output:
Unique:    240755312
Distinct:  821667195
Total:     40889468950
Max_count: 11354815

# After obtaining the results, we need to convert the results into a histogram, namely a .histo file, for subsequent downstream analysis using GenomeScope

jellyfish histo -t 16 my_kmer21.jf > my_kmer21.histo

# Next, we will introduce how to perform subsquent GenomeScope visualization analysis
# We first fit a model and generate outputs for the jellyfish histogram we just created

Rscript genomescope.R -i my_kmer21.histo -o genomescope_result -k 21

# Possible results:

GenomeScope analyzing my_kmer21.histo p=2 k=21 outdir=genomescope_result
# Fit my_kmer21.histo with ploidy p=2 (diploid) and k=21, and save the results to the genomescope_result directory
aa:98.2% ab:1.75%
# aa: Proportion assigned to the homozygous k-mer component
# ab: Proportion assigned to the heterozygous k-mer component
Model converged het:0.0175 kcov:23.4 err:0.000815 model fit:0.979 len:700799039

# View and save the visualization results
ls genomescope_result/

# ENDING !!!




```
